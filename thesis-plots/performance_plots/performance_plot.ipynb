```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator

df = pd.read_csv('results.csv')

# Plotting the data points
plt.plot(df["Items"], df["Total Time"], marker='o', linestyle='None', label='Measured Times')

z = np.polyfit(df["Items"], df["Total Time"], 1)
p = np.poly1d(z)
plt.plot(df["Items"], p(df["Items"]), "r--", label='Linear Trendline')

plt.xlabel("Amount of Items")
plt.ylabel("Total Time (s)")
plt.title("Plot of Total Time vs. Amount of Items")

plt.gca().xaxis.set_major_locator(MaxNLocator(6))

plt.legend()
plt.savefig('performance_plot.pdf', bbox_inches='tight')
```
